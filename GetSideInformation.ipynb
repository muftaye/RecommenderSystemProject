```python
genres=[-1]*len(games)
popularities=[-1]*len(games)
ratings=[-1]*len(games)
themes=[-1]*len(games)
missingData=[]
notFound=[]
ind=2187
for indx,game in enumerate(games[2187:]):
    result = igdb.games({
        'filters' :{
            "[name][eq]": game
        },
        'fields' : fields,
    })
    #print result.body
    if(result.body!=[]):
        try:
            genres[ind]=result.body[0]['genres']
            popularities[ind]=result.body[0]['popularity']
            ratings[ind]=result.body[0]['rating']
            themes[ind]=result.body[0]['themes']
        except:
            print ind, '--------------------has missing data'
            missingData.append(ind)
    else:
        result = igdb.games({
            'search': game,
            'fields' : fields,
        })
        if(result.body!=[]):
            try:
                genres[ind]=result.body[0]['genres']
                popularities[ind]=result.body[0]['popularity']
                ratings[ind]=result.body[0]['rating']
                themes[ind]=result.body[0]['themes']
            except:
                print ind, '--------------------has missing data'
                missingData.append(ind)
        else:
            print '--------------------cannot find:',game,":",ind
            notFound.append(ind)
    ind+=1
print 'done'
```
